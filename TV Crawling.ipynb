```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://workey.codeit.kr/orangebottle/index"
res = requests.get(url)
soup = bs(res.text, "html.parser")
branches = soup.select("div.branch")
branch_infos = []
for branch in branches:
    branch_name = branch.select_one("p.city").text
    address = branch.select_one("p.address").text
    phone_number = branch.select_one("span.phoneNum")
    branch_infos.append([branch_name,address,phone_number])
branch_infos
```

```python
# find, findall로 사용해보기 
```

```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://workey.codeit.kr/orangebottle/index"
res = requests.get(url)
soup = bs(res.text, "html.parser")
branches = soup.find_all("div", class_="branch") 

result = []
for branch in branches:
    name = branch.find("p", {"class":"city"}).get_text()
    address = branch.find("p",{"class":"address"}).get_text()
    phone = branch.find("span",{"class":"phoneNum"}).get_text()
    result.append([name, address, phone])
result
```

```python
import csv 
import requests
from bs4 import BeautifulSoup as bs 

# csv 파일생성
file = open("C:\\Users\\yunaj\\test\\시청률.csv","w", encoding="utf-8-sig")
csv_writer = csv.writer(file)

# 헤더 행 추가
csv_writer.writerow(["순위",'채널', '프로그램', '시청률'])

res = requests.get("https://workey.codeit.kr/ratings/index")
soup = bs(res.text, "html.parser")
for tr_tag in soup.select('tr')[1:]:
    td_tags = tr_tag.select('td')
    row = [
            td_tags[0].get_text(),
            td_tags[1].get_text(),
            td_tags[2].get_text(),
            td_tags[3].get_text()
    ]
    csv_writer.writerow(row)
file.close()
```

```python
# 연도별 시청률 데이터 엑셀 파일로 정리해 보기
```

```python
import requests
from bs4 import BeautifulSoup as bs
from openpyxl import Workbook 


wb = Workbook(write_only=True) 
for year in range(2010,2019):
    ws = wb.create_sheet(f"{year}년")
    ws.append(["기간","순위","채널","프로그램","시청률"])
    for month in range(1,13):
        for week in range(0,5):
            url = f"https://workey.codeit.kr/ratings/index?year={year}&month={month}&weekIndex={week}"
            res = requests.get(url)
            soup = bs(res.text, "html.parser")
           
            for tag in soup.select("tr")[1:]:                 
                period = "{}년 {}월 {}주차".format(year, month, week+1)
                row=[
                    period,
                    tag.select_one("td.rank").get_text(),
                    tag.select_one("td.channel").get_text(),
                    tag.select_one("td.program").get_text(),
                    tag.select_one("td.percent").get_text()
                ]
                ws.append(row)
wb.save("C:\\Users\\yunaj\\test\\시청률2.xlsx")     
```

```python
import csv
import requests
from bs4 import BeautifulSoup as bs


# csv 파일 생성
file = open("C:\\Users\\yunaj\\test\\시청률final.csv","w",encoding="utf-8-sig")
csv_writer = csv.writer(file)

# 헤더행
csv_writer.writerow(["순위",'채널', '프로그램', '시청률'])

for year in range(2016,2019):
    for month in range(1,13):
        for week in range(0,5):
            url = f"https://workey.codeit.kr/ratings/index?year={year}&month={month}&weekIndex={week}"
            res = requests.get(url)
            soup = bs(res.text, "html.parser")
            contents = soup.select("tr.row")[1:]             
            for x in contents:
                if x.select_one("td.channel").get_text() == "SBS": 
                    row = [
                        x.select_one("td.rank").text,
                        x.select_one("td.channel").text,
                        x.select_one("td.program").text,
                        x.select_one("td.percent").text
                    ]
                    csv_writer.writerow(row)
                       
file.close()
    
    
```
